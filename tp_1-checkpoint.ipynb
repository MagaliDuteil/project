```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns
```

```python
df_auto = pd.read_csv('C:/Users/Magal/OneDrive/Bureau/cleaned_autos.csv')
df_auto = pd.DataFrame(df_auto)
```

```python
# - Nombre total de véhicules en vente selon le type de véhicule.
```

```python
df_veh_tot_type = df_auto.groupby('vehicleType').size().sort_values(ascending = False)
print(df_veh_tot_type)
```

```python
plt.bar(df_veh_tot_type.index,df_veh_tot_type, color = '#02FF88')
plt.xticks(rotation = 45)

```

```python
#Répartition des véhicules en fonction de l'année d'immatriculation
```

```python
df_veh_ann_immat = df_auto.groupby('yearOfRegistration').size().sort_values(ascending=False)
print(df_veh_ann_immat)
```

```python
filtered_years = df_auto['yearOfRegistration'][(df_auto['yearOfRegistration'] >= 1980) & (df_auto['yearOfRegistration'] <= 2023)]
plt.figure(figsize=(12, 6))
plt.hist(filtered_years, bins=20, edgecolor='k')
plt.xlabel('Année de immatriculation')
plt.ylabel('Nombre de Véhicules')
plt.title('Véhicules par Année immatriculation (depuis 1980)')
plt.xticks(range(1980, 2024, 5))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#Nombre de véhicules par marque
```

```python
df_veh_tot_marque = df_auto.groupby('brand').size().sort_values(ascending = False)
print(df_veh_tot_marque)
```

```python
plt.figure(figsize=(12, 6))
df_veh_tot_marque.plot(kind='barh', color='skyblue')
plt.xlabel('Nombre de Véhicules')
plt.ylabel('Marque')
plt.title('Nombre de Véhicules par Marque')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Prix moyen des véhicules par type de véhicule et type de boîte de vitesses
```

```python
df_aut_prix_mean = df_auto.groupby(['vehicleType', 'gearbox'])['price'].mean().reset_index()
df_aut_prix_mean = df_aut_prix_mean.sort_values(by='price', ascending=False)
print(df_aut_prix_mean)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(range(len(df_aut_prix_mean)), df_aut_prix_mean['price'], tick_label=df_aut_prix_mean['vehicleType'] + ' - ' + df_aut_prix_mean['gearbox'])
plt.xlabel('Moyenne de prix par type et boite de vitesse')
plt.ylabel('Moyenne des prix')
plt.title('Type véhicule et ')
plt.xticks(rotation=45, ha='right')
```

```python
df_aut_prix_mean_carbu = df_auto.groupby(['fuelType', 'gearbox'])['price'].mean().reset_index()
df_aut_prix_mean_carbu = df_aut_prix_mean_carbu.sort_values(by='price', ascending=False)
print(df_aut_prix_mean)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(range(len(df_aut_prix_mean_carbu)), df_aut_prix_mean_carbu['price'], tick_label=df_aut_prix_mean_carbu['fuelType'] + ' - ' + df_aut_prix_mean_carbu['gearbox'], color='#FFAF02')
plt.xlabel('Moyenne de prix par carburant et boite de vitesse')
plt.ylabel('Moyenne des prix')
plt.title('Carburant et boite de vitesse')
plt.xticks(rotation=45, ha='right')
```

```python
df_aut_marque_top = df_auto.groupby(['brand', 'vehicleType'])['price'].mean().reset_index()
df_aut_marque_top = df_aut_marque_top.sort_values(by='price', ascending=False)
df_aut_marque_top.head(10)
```

```python
heatmap_data = df_aut_marque_top.pivot_table(index='brand', columns='vehicleType', values='price')

# Créer un heatmap avec Seaborn
plt.figure(figsize=(12, len(heatmap_data) * 0.20))
sns.heatmap(heatmap_data, cmap='rocket_r', annot=True, fmt=".2f", linewidths=0.5)

plt.xlabel('Marque')
plt.ylabel('Type de véhicule')
plt.title('Heatmap des moyennes de prix par marque et type de véhicule')

# Ajuster la rotation des étiquettes
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

plt.tight_layout()
plt.show()
```
